```python
import textract
import os

def convert_doc_to_text(input_directory, output_directory):
    # Ensure the output directory exists; if not, create it
    if not os.path.exists(output_directory):
        os.makedirs(output_directory)
    
    for filename in os.listdir(input_directory):
        if filename.endswith(".doc"):
            input_path = os.path.join(input_directory, filename)
            output_filename = os.path.splitext(filename)[0] + ".txt"
            output_path = os.path.join(output_directory, output_filename)           
            text = textract.process(input_path)
            decoded_text = text.decode('utf-8')
            with open(output_path, "w", encoding="utf-8") as file:
                file.write(decoded_text)
                
            print(f"Converted {input_path} to {output_path}")

```

```python
def flatten_and_remove_spaces(input_filename, output_filename):
    try:
        with open(input_filename, 'r') as input_file:
            # c ni dung t tp v loi b du xung dng
            content = input_file.read().replace('\n', ' ')
            # Tch cc t t ni dung
            words = content.split()
            # Ghi cc t li thnh mt dng mi m khng c khong trng gia chng
            flattened_content = ' '.join(words)
            # Ghi ni dung  c bin i vo tp mi
            with open(output_filename, 'w') as output_file:
                output_file.write(flattened_content)
        print(" ghi thnh cng ni dung  bin i vo '{}'.".format(output_filename))
    except FileNotFoundError:
        print("Khng tm thy tp '{}'.".format(input_filename))

# Thay 'input_file.txt' bng tn ca tp bn mun c
input_filename = '/home/justtuananh/AI4TUAN/VA/PhuNu?/data/converted/1.txt'
# Thay 'output_file.txt' bng tn ca tp bn mun ghi ni dung  bin i vo
output_filename = '1.txt'
flatten_and_remove_spaces(input_filename, output_filename)
```

```python
import re

def get_chapters(text):
  """
  Ly danh sch cc chng t on vn bn.

  Args:
    text: on vn bn cn x l.

  Returns:
    Danh sch cc chng (mi chng l mt chui).
  """
  # Biu thc regex  tm tn chng
  regex = r"(?<=CHNG )([IVX]+)(?= )"

  # Tm kim tn chng trong vn bn
  chapters = []
  for match in re.finditer(regex, text):
    chapter_number = match.group(1)
    chapter_name = get_chapter_name(text[match.start():match.end()])
    chapters.append(f"{chapter_number}. {chapter_name}")
  return chapters
```

```python
def extract_content_before_first_chapter(text):
    match = re.search(r'(?:Chng|CHNG)\s+([IVXLCDM]+)(?:\s|$)', text, re.IGNORECASE)
    if match:

        content_before_first_chapter = text[:match.start()].strip()
        return content_before_first_chapter
    else:
        return None

    
def extract_abstract(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        text = file.read()
        content = extract_content_before_first_chapter(text)
        return content
```

```python
from typing import List, Optional
import os
from langchain.docstore.document import Document
from langchain_text_splitters import RecursiveCharacterTextSplitter
EMBEDDING_MODEL_NAME = "bkai-foundation-models/vietnamese-bi-encoder"


def split_documents(
    chunk_size: int,
    knowledge_base: List[Document],
    tokenizer_name: Optional[str] = EMBEDDING_MODEL_NAME,
) -> List[Document]:
    """
    Split documents into chunks of maximum size `chunk_size` tokens and return a list of documents.
    """
    text_splitter = RecursiveCharacterTextSplitter.from_huggingface_tokenizer(
        AutoTokenizer.from_pretrained(tokenizer_name),
        chunk_size=chunk_size,
        chunk_overlap=int(chunk_size / 10),
        add_start_index=True,
        strip_whitespace=True,
    )

    docs_processed = []
    for doc in knowledge_base:
        doc.metadata['source'] = os.path.splitext(os.path.basename(doc.metadata['source']))[0]
        docs_processed += text_splitter.split_documents([doc])

    # Remove duplicates
    unique_texts = {}
    docs_processed_unique = []
    for doc in docs_processed:
        if doc.page_content not in unique_texts:
            unique_texts[doc.page_content] = True
            docs_processed_unique.append(doc)

    return docs_processed_unique
```
